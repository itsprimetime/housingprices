````python
y = df['sale_price']    # what we want to predict (ie. price)
X = df['lot_area']      # feature
````

````python
X = sm.add_constant(X)
X.head()
````

##### In StatsModels we have to manually add the y-intecept by hand by using the sm.add_constant() command.

````python
# First we instantiate our model
model = sm.OLS(y, X)
````

````python
# Then we fit our model to the data
# The process of fiting our model means calculating B_naught and Beta_one

model = model.fit()
````

````python
predictions = model.predict(X)
````

````python
model.summary()
````

### Interpretation of the Slope and Intercept

##### Our Beta_Naught is 146,400. This means that if lot area is 0, then we can expect the price of the home to be $146,000.

##### Our Beta_One is 3.28. This means that for every one unit change in lot_area, we can expect the sale price to increase by $3.48.

##### That said, we cannot use this model because there is strong multicolinearity.

### Lasso

````python
from sklearn.linear_model import Lasso
````

````python
def lasso_coefs(X, y, alphas):
    coefs = []
    lasso_reg = Lasso()
    for a in alphas:
        lasso_reg.set_params(alpha=a)
        lasso_reg.fit(X, y)
        coefs.append(lasso_reg.coef_)
    return coefs
````

````python
l_alphas = np.arange(0.001, 0.15, 0.0025)
l_coefs = lasso_coefs(X, y, l_alphas)
````

````python
def lasso_plot_runner(alpha=0):
    coef_plotter(l_alphas, l_coefs, X.columns, alpha, regtype='lasso')

interact(lasso_plot_runner, alpha=(0.001,0.2,0.01))
````

## Model Prep (Train / Test Split)

````python
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=42)
````

## Model Prep: Scaling

````python
ss = StandardScaler()
ss.fit(X_train)
X_train_sc = ss.transform(X_train)
X_test_sc = ss.transform(X_test)
````

## Model Prep: Instantiate our models

````python
lr = LinearRegression()
````

````python
lasso = LassoCV(n_alphas=100)
````

````python
ridge = RidgeCV(alphas=np.linspace(.1, 10, 100))
````

## Cross Validation

````python
lr_scores = cross_val_score(lr, X_train_sc, y_train, cv=3)
lr_scores.mean()
````

````python
lasso_scores = cross_val_score(lasso, X_train_sc, y_train, cv=3)
lasso_scores.mean()
````

````python
ridge_scores = cross_val_score(ridge, X_train_sc, y_train, cv=3)
ridge_scores.mean()
````
